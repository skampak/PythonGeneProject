```python
import pandas as pd
gencode = pd.read_table("gencode.v29.annotation.gff3", comment="#", sep = "\t", names = ['seqname', 'source', 'feature', 'start' , 'end', 'score', 'strand', 'frame', 'attribute'])
gencode=gencode[gencode.source != 'HAVANA']
gencode.head() 
gencode.info()
gencodeexin=gencode.drop(['source','score','strand','frame'], axis=1)
```

```python
def get_transcript(x):
    z  = dict([y.split('=') for y in x.split(';')])
    return z.get('transcript_id')

def get_tag(x):
    z  = dict([y.split('=') for y in x.split(';')])
    return z.get('tag')

def get_genename(x):
    z  = dict([y.split('=') for y in x.split(';')])
    return z.get('gene_name')

def get_geneid(x):
    z  = dict([y.split('=') for y in x.split(';')])
    return z.get('gene_id')
```

```python
gencodeexin['transcript'] = gencodeexin.apply(lambda x: get_transcript(x['attribute'])  , axis=1)
gencodeexin['tag'] = gencodeexin.apply(lambda x: get_tag(x['attribute'])  , axis=1)
gencodeexin['geneid'] = gencodeexin.apply(lambda x: get_geneid(x['attribute'])  , axis=1)
gencodeexin['genename'] = gencode.apply(lambda x: get_genename(x['attribute'])  , axis=1)
```
